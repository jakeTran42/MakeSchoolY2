```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats

from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC

from sklearn.preprocessing import StandardScaler
from sklearn.metrics import confusion_matrix
from sklearn import metrics
from sklearn.metrics import recall_score, precision_score, f1_score, auc

# use seaborn plotting defaults
import seaborn as sns; sns.set()
```

```python
cancer = load_breast_cancer()
```

```python
print(cancer.DESCR)
```

```python
data = pd.DataFrame(cancer.data)
data.columns = cancer.feature_names
# data["Target"] = cancer.target
```

```python
data.head()
```

```python
# data.drop(data.columns[[-1, 0]], axis=1, inplace=True)
# data.head()
```

```python
features_mean= list(data.columns[1:11])
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(data[features_mean].corr(), annot=True, square=True, cmap='coolwarm')
plt.show()
```

```python
scaler = StandardScaler()
```

```python
cancer_df = pd.DataFrame(data=scaler.fit_transform(data), columns=cancer.feature_names)
cancer_target = pd.Series(data=cancer.target)
```

```python
cancer_df.describe()
```

```python
X_train, X_test, y_train, y_test = train_test_split(cancer_df, cancer_target, test_size = .2)
```

```python
model = SVC()
lm = model.fit(X_train, y_train)
```

```python
confusion_matrix(y_test, lm.predict(X_test))
```
